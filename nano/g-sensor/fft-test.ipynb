```python
import numpy as np
import matplotlib.pyplot as plt

# create a sinus array with 1000 samples
x = np.arange(0, 1024)
y = np.sin(x * 2 * np.pi / 80)

# get freq response of the signal
yf = np.fft.rfft(y)

xf = np.fft.rfftfreq(len(y), 1/1000)
print (xf)
# create a figure and a subplot
fig, ax = plt.subplots()

# plot the fft
ax.plot(xf,yf)


# plot the data
ax.plot(x, y)



# show the plot
plt.show()
```
